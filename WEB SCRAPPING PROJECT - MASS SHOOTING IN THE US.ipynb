```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_mass_shootings_in_the_United_States'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_='wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
doc_titles = table.find_all('th') 
```

```python
doc_titles
```

```python
doc_table_titles = [title.text.strip() for title in doc_titles]
print(doc_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = doc_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
    
    
  
```
